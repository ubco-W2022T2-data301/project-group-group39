```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../data/raw/archive (2)/all_seasons.csv")
grouped = df.groupby('season')
top10_avg = grouped.apply(lambda x: x.nlargest(10, 'pts').mean(numeric_only=True)).reset_index()[['season', 'pts']] 
numeric_only=None
top10_avg.columns = ['season', 'avg_top10_pts']
sns.set(style="darkgrid")
sns.catplot(y='season', x='avg_top10_pts', kind='bar', data=top10_avg, color='grey')


```

* with this graph it looks like it's pretty consistent where it average from 25- 30

* from 2010-2011 to 2013-2014 seems to be an era of lower scoring

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
top10 = df.groupby('season').apply(lambda x: x.nlargest(10, 'pts')).reset_index(drop=True)
fig, ax = plt.subplots()
ax.scatter(top10['season'], top10['pts'])
ax.set_xlabel('season')
ax.set_ylabel('pts')
ax.set_title('Top scorers per Season')

ax.set_xticklabels(['1996', '1997', '1998', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', "2006","2007","2008","2009","2010","2011","2012","2013","2014","2015", "2016","2017","2018","2019","2020","2021" ])
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_yticklabels([0, 5,10,15,20,25,30,35,40])
plt.show()

```

* you can see two dots that near 40 pts in 2004 and 2017 which bring up the average for those year a lot  

* you can see from 2011 and on, the lowest pts from the top 10 scorers of each season start to increase in a linear line, thus having the average going up

* in the newer seasons, the dots seem to be closer with one another then the ealier years. thsu players are scoring closer to each other then before

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/archive (2)/all_seasons.csv")
avg_top_scorers = df.groupby('season')['pts'].nlargest(10).groupby('season').mean()
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(avg_top_scorers.index, avg_top_scorers.values)
ax.set_xlabel('Year')
ax.set_ylabel('Average Points per Game')
ax.set_title('Average of the Top 10 Scorers per Season')
tick_labels = [str(year) for year in avg_top_scorers.index]
ax.set_xticklabels(tick_labels, rotation=90)

plt.show()


```

* can see the bigger difference between each season which is good because even one point is a big difference between one another

* 2005-2006 and 2019-2020 were the peaks of the graph, which is 14 years between each other. Big age difference but not huge pts difference

* 1997-1998 was the lowest average pts by a whole point
